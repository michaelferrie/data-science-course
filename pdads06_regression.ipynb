#### Equation of a line

We can represent lines by the following equation:

$$ y = mx + b $$

In this equation m represents the slope or gradient of the line and is a coefficient or multiplier of the value of x, b (sometimes written as c) is the point at which the line crosses the y axis, this is known as the y-intercept. In other words, m, represents rate of change and the b represents the starting point of the line as it crosses the y axis.

Consider the following illustration:

image of line graph


We can see that the y intercept is at 0 so we can write this as: $ y = mx + 0 $ or just $ y = mx $. The question is, how do we find m? The slope (m) requires some calculation. To calculate m we divide the rise of the line by the run, how far up over how far along. Rise is how far up, run is how far along.

Consider our example graph again, the line rises up by two points on the y axis and runs by one point on the x axis, so the formula for this would be:

Rise / Run 
= 2 / 1 
= 2

The slope of the line (m) is 1, so the formula of the line from illustration 1 is $ y = 2x + 0 $, this can actually be written as just simply $ y = 2x $. Let us look at another example.


#### Statistics terminology

In statistics they use a different version of this equation, instead of 

$$ y = mx + b $$

In statistics we put the y intercept first and use B0 for this and the slope is called B1. The slope is called the regression coefficient. They call these B’s beta, so a statistics person might say, "Y= Beta 0 plus Beta 1 x," which is the same as first equation in the notebook.

$$ y = \beta0 + \beta1x $$

When we have two sets of points in a data set, we can consider one of them being the dependent variable (y) and one of them being the independent variable (x). So The change in one depends on the change in the other. One of them is dependent on the other in terms of what degree of change it makes.

#### Adding a best fit line

Now that we can see the data on a graph, we can add a best fit line to the data, this can be achieved by using the `polyfit` function of numpy.

```python
# Import and alias the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# define two arrays
height = np.array([65, 65, 62, 67, 69, 65, 61, 67])
weight = np.array([105, 125, 110, 120, 140, 135, 95, 130])

# find line equation, beta1 (coefficient) and beta0 (y-intercept)
beta1, beta0 = np.polyfit(height, weight, 1)

# print beta values
print("The y-intercept is: ", beta0)
print("The slope of the line is: ", beta1)
print("The equation of the line is")
print("y = ", beta0, '+', beta1, '* x')
```

```python
# create scatter plot
plt.scatter(height, weight)

# add best fit line using formula
plt.plot(height, beta0 + beta1 * height)         

# show the plot
plt.show()
```

Add our labels back into the plot

```python
# Import and alias the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# define two arrays
height = np.array([65, 65, 62, 67, 69, 65, 61, 67])
weight = np.array([105, 125, 110, 120, 140, 135, 95, 130])

# find the best fit line, create beta0 and beta1
beta1, beta0 = np.polyfit(height, weight, 1)

# create scatter plot
plt.scatter(height, weight)

# add best fit line using formula
plt.plot(height, beta0 + beta1 * height)

# add labels
plt.title('Plant data scatter plot, with best fit line')
plt.ylabel('x-axis - Weight in grams')
plt.xlabel('y-axis Height in cm')

# show the plot
plt.show()
```

#### Simple linear regression

Returning to our height and weight dataset, there was one final value did not have a height, data point 9, which has a weight of 71 grams, however now that we have this best fit trend line we should be able to calculate where this point would fit, based on the data we have.

We know how to generate the linear equation, we can just pass values into the formula and can predict based on the equation of the line what the value would be. We have x=71 and we should be able to predict y.

$$ y = \beta0 + \beta1x $$

We know the slope of the line \beta1 and we know the y-intercept \beta0 so we can just pass in another value to the equation to get the result of y. 

```python
# define weight of plant number 9 as x
x = 71

# create a variable and pass in x to equation of the line
plant_9 = beta0 + (beta1 * x)
print("According to best fit line this would be: ", plant_9 , "cm")
```

That's great, once we have the equation of the line, we can predict other y values given any x value, according to out data what if a plant was 100 grams? Let's see...

```python
# define weight of plant number 10 as x
x = 100

# create a variable and pass in x to equation of the line
plant_9 = beta0 + (beta1 * x)
print("According to best fit line this would be: ", plant_9 , "cm")
```

#### Examples where data doesn't suit regression

Although this is a very small dataset, it would not be large enough to make a proof of concept, ideally if you are surveying a group of people you want to gather as much data as possible. Let's compare the age and weight lost on the diet. 

First look at those two columns, then we will create two numpy arrays with them to generate a plot, note - if we print more than one column we need to use double square brackets.

We can use `to_numpy()` to take columns out and make them arrays.

```python
# print columns
print(df[['Age', 'pre.weight', 'weight6weeks']])

# use to.numpy()
age = df['Age'].to_numpy()
pre = df['pre.weight'].to_numpy()
post = df['weight6weeks'].to_numpy()
print(age, pre, post)
```

This is ideal we have 3 arrays now but actually what we are looking for is the total weight loss, so we need to subtract the post diet weight from the pre diet weight. It is probably easier to do this in pandas first.

This is a great pandas feature we can just add a new column to the dataframe and tell it to calculate this based on the result of some calculation on two other columns.

```python
# define a new column as the result of a calculation on two others
df['weight.loss'] = df['pre.weight'] - df['weight6weeks']
print(df)

# make a new array of total weight loss
loss = df['weight.loss'].to_numpy()
print(loss, age)
```

#### Visualise the data

Create a plot from the data, a stronger correlation is better for a regression, check the correlation first.

There is a very weak negative positive correlation, this will make this analysis more challenging, generally there needs to be a trend to be able to draw a line.

```python
# import and alias the necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# check correlation
print(np.corrcoef(age, loss)[0,1])

# plot data
plt.scatter(loss, age)
```

Try to compare the height with the weight loss, again a very weak association. Let's see what the best fit line looks like.

```python
# create height array
height = df['Height'].to_numpy()

# plot data
plt.scatter(loss, height)

# print correlation
print(np.corrcoef(loss, height)[0,1])
```

```python
# create scatter plot
plt.scatter(loss, height)

#find line of best fit
beta0, beta1 = np.polyfit(loss,height, 1)

# add best fit line using formula
plt.plot(loss, beta0 * loss + beta1)

# add labels
plt.title('Weight loss and height')
plt.xlabel('x-axis - Kgs')
plt.ylabel('y-axis Height in cm')

# show the plot
plt.show()
```

Often we want to select one column based on an entry in another, or only pull a specific row in pandas. The following are examples on how to do this, first using index numbers.

Search for a specific string using `str.contains`

`df['Column Name'].str.contains('String')`

```python
# check for a match in a column
df['region'].str.contains("northwest")
```

String match will only return the values that match, what if we want to match on two columns. To look for numeric values, best to use `isin`, the number needs to go in square brackets.

These is a [useful reference](https://www.w3schools.com/python/pandas/pandas_ref_dataframe.asp) of more of these methods of the dataframe object with examples on the W3 website.

```python
# str.match on one column
df[df['region'].str.match('northwest')]

# str.match for only northeast
new_df = df[df['region'].str.match('northwest')]

# only people with 1 child
new_df[new_df['children'].isin([1])]
```

We can also search or more two columns at the same time using & for AND and | for OR. I found a good guide for this [here](https://www.ritchieng.com/pandas-multi-criteria-filtering/).

```python
# search on two columns at the same time with AND & - in northwest and a smoker
df[df['region'].str.match('northwest') & df['smoker'].str.match('no')]

# search on two columns using OR | either in northwest or a smoker
df[df['region'].str.match('northwest') | df['smoker'].str.match('yes')]

# three columns, you can keep nesting them...
df[df['region'].str.match('northwest') & df['smoker'].str.match('yes') & df['children'].isin([1])]
```
